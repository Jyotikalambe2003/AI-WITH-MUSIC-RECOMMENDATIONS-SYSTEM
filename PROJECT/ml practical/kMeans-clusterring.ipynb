```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
import numpy as np
```

```python
np.random.seed(18)
c1 = np.random.randint(0, 100, 50)
c2 = np.random.randint(0, 100, 50)
print(c1)
print(c2)
```

```python
data = {'col1': c1, 'col2': c2}
df = pd.DataFrame(data)
```

```python
df
```

```python
km=KMeans(n_clusters=3,random_state=24)
km
```

```python
y_predict=km.fit_predict(df[['col1','col2']] )
y_predict
```

```python
df["Cluster"]=y_predict
```

```python
df
```

```python
centroids = km.cluster_centers_
df['Cluster'] = km.labels_
```

```python
plt.scatter(df["col1"],df['col2'])
```

```python
# Visualize the clusters
plt.scatter(df['col1'], df['col2'], c=df['Cluster'], cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200, label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('col1')
plt.ylabel('col2')
plt.legend()
plt.show()
```
