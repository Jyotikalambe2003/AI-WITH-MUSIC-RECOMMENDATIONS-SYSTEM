```python
import matplotlib.pyplot as plt
import pandas as pd
# import iris data set
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
# gettinf data set
iris=load_iris()
```

```python
# data frme
df=pd.DataFrame(iris.data)
```

```python
df
```

```python
df.columns=['sepal length',
 'sepal width',
 'petal length',
 'petal width']
```

```python
df
```

```python
# number of cluster:
k = 3  
```

```python
# calling kmeans
kmeans = KMeans(n_clusters=k)
```

```python
# fit 
kmeans.fit(df)

```

```python
# new column cluster
df['cluster'] = kmeans.fit_predict(df)
```

```python
df
```

```python
# creating cluster center
centroids = kmeans.cluster_centers_
```

```python
# centoroid calculation
df_centroids = pd.DataFrame(centroids, columns=['sepal length', 'sepal width', 'petal length', 'petal width'])
df_centroids['cluster'] = ['Centroid'] * k 
```

```python
# adding cluster centroid :
df = pd.concat([df, df_centroids], ignore_index=True)
```

```python
df
```

```python
# display

plt.scatter(df[df['cluster'] != 'Centroid']['sepal length'], df[df['cluster'] != 'Centroid']['sepal width'],
            c=df[df['cluster'] != 'Centroid']['cluster'], cmap='viridis', marker='o', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=100, label='Centroids')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-Means Clustering with Centroids')
plt.legend()
plt.show()



```
