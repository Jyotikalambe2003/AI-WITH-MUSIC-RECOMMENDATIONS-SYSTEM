```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# loading image
image = cv2.imread('dog.jpg')  # Replace 'dog.jpg' with your image path
```

```python
image
```

```python
# Rotate angle
rotation_angle = 30 
# rotation time
rotation_steps = 12 
```

```python
for i in range(rotation_steps):
    # Create a martrix rotation:
    M = cv2.getRotationMatrix2D((image.shape[1] / 2, image.shape[0] / 2), i * rotation_angle, 1)

    # apply rotation
    rotated_image = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))

    plt.subplot(4, 3, i + 1)
    plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB)) 
    if i == 0:
        plt.title('Rotation')
    if i == rotation_steps - 1:
        plt.xlabel(f'{i * rotation_angle} degrees')

plt.show()
```

```python

# display layout
plt.figure(figsize=(12, 6))
```

```python
# Adjust brightness 
brightness = -25  
# Adjust contrast 
contrast = 0.5


```

```python
# Apply brightness and contrast adjustments
adjusted_image = cv2.convertScaleAbs(image, alpha=contrast, beta=brightness)
```

```python

plt.figure(figsize=(8, 4))
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(122), plt.imshow(cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB)), plt.title('Adjusted')
plt.show()
```

```python
# resizing image
new_width = 300
new_height = 200

resized_image = cv2.resize(image, (new_width, new_height))
```

```python
# display
plt.figure(figsize=(8, 4))
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(122), plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)), plt.title('Resized')
plt.show()
```
